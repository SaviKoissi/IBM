```python
import pandas as pd
import requests

# Make a GET request to the SpaceX API
url = "https://api.spacexdata.com/v4/launches"
response = requests.get(url)
data = response.json()

# Convert the response to a DataFrame using pd.json_normalize
df = pd.json_normalize(data)

# Print unique values in the 'rocket' column
unique_rockets = df['rocket'].unique()
print("All Rockets:")
for rocket in unique_rockets:
    print(rocket)
```

```python
# Global variables 
BoosterVersion = []
PayloadMass = []
Orbit = []
LaunchSite = []
Outcome = []
Flights = []
GridFins = []
Reused = []
Legs = []
LandingPad = []
Block = []
ReusedCount = []
Serial = []
Longitude = []
Latitude = []

# Function to get booster information
def get_booster_info(booster_id):
    global BoosterVersion
    if booster_id:  # Check if booster_id is not None
        response = requests.get(f'https://api.spacexdata.com/v4/rockets/{booster_id}').json()
        if response:
            booster_data = response.get('name')
            if booster_data:  # Check if booster_data is not None
                BoosterVersion.append(booster_data)
            else:
                BoosterVersion.append(None)
        else:
            BoosterVersion.append(None)
    else:
        BoosterVersion.append(None)


# Function to get payload information
def get_payload_info(payload_id):
    global PayloadMass, Orbit
    payload_url = f'https://api.spacexdata.com/v4/payloads/{payload_id}'
    response = requests.get(payload_url)
    if response.status_code == 200:
        payload_data = response.json()
        PayloadMass.append(payload_data.get('mass_kg'))
        Orbit.append(payload_data.get('orbit'))
    else:
        PayloadMass.append(None)
        Orbit.append(None)

# Function to get launchpad information
def get_launchpad_info(launchpad_id):
    global LaunchSite, Longitude, Latitude
    launchpad_url = f'https://api.spacexdata.com/v4/launchpads/{launchpad_id}'
    response = requests.get(launchpad_url)
    if response.status_code == 200:
        launchpad_data = response.json()
        LaunchSite.append(launchpad_data.get('name'))
        Longitude.append(launchpad_data.get('longitude'))
        Latitude.append(launchpad_data.get('latitude'))
    else:
        LaunchSite.append(None)
        Longitude.append(None)
        Latitude.append(None)

# Function to get core information
def get_core_info(core_id):
    global Outcome, Flights, GridFins, Reused, Legs, LandingPad, Block, ReusedCount, Serial
    core_url = f'https://api.spacexdata.com/v4/cores/{core_id}'
    response = requests.get(core_url)
    if response.status_code == 200:
        core_data = response.json()
        Outcome.append(core_data.get('landing_outcome'))
        Flights.append(core_data.get('flights'))
        GridFins.append(core_data.get('gridfins'))
        Reused.append(core_data.get('reused'))
        Legs.append(core_data.get('legs'))
        LandingPad.append(core_data.get('landpad'))
        Block.append(core_data.get('block'))
        ReusedCount.append(core_data.get('reuse_count'))
        Serial.append(core_data.get('serial'))
    else:
        Outcome.append(None)
        Flights.append(None)
        GridFins.append(None)
        Reused.append(None)
        Legs.append(None)
        LandingPad.append(None)
        Block.append(None)
        ReusedCount.append(None)
        Serial.append(None)

# Loop through the dataframe and call functions
for index, row in df.iterrows():
    get_booster_info(row['rocket'])
    get_payload_info(row['payloads'])
    get_launchpad_info(row['launchpad'])
    get_core_info(row['cores'])

# Create a new dataframe
new_data = pd.DataFrame({
    'BoosterVersion': BoosterVersion,
    'PayloadMass': PayloadMass,
    'Orbit': Orbit,
    'LaunchSite': LaunchSite,
    'Outcome': Outcome,
    'Flights': Flights,
    'GridFins': GridFins,
    'Reused': Reused,
    'Legs': Legs,
    'LandingPad': LandingPad,
    'Block': Block,
    'ReusedCount': ReusedCount,
    'Serial': Serial,
    'Longitude': Longitude,
    'Latitude': Latitude
})

```
